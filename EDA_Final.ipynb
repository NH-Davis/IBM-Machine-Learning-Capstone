### Task: Find the Top-20 Most Popular Courses


_TODO: Use Pandas groupby() and size() methods on the item column to aggregate the rating count for each item, then use the sort_values() method to sort the course enrollment count, and use the slice method to get the top 20 courses. You may also implement this task with different solutions_


```python
import pandas as pd

# Group by 'item' and count the number of ratings for each course
course_rating_counts = ratings_df.groupby('item').size()

# Sort the counts in descending order to find the top 20 most rated courses
top_courses = course_rating_counts.sort_values(ascending=False).head(20)

# Convert this to a DataFrame
top_courses_df = top_courses.reset_index()
top_courses_df.columns = ['item', 'rating_count']
```

_TODO: Use Pandas merge() method to join the course_df (contains the course title column)._


```python
from IPython.display import display  # Import the display function

# Perform the merge to include the 'TITLE' from course_df
merged_details = pd.merge(top_courses_df, course_df, how='left', left_on='item', right_on='COURSE_ID')

# Selecting only the necessary columns to display: item (course ID from top_courses_df), rating_count, and TITLE
final_df = merged_details[['rating_count', 'TITLE']]

# Display the merged DataFrame with course titles
print(final_df)
```
